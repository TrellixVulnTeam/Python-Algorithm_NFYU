```python
from collections import deque

def bfs(n):
    q = deque()
    q.append(n)
    while q:
        x = q.popleft()
        if x == m:
            print(dist[x])
            
        for i in (x-1, x+1, x*2):
            if 0<= i <= MAX and not dist[i]:
                dist[i] = dist[x] + 1
                q.append(i)
                
MAX = 10 ** 5
n, m = map(int, input().split())
dist = [0] * (MAX + 1)
cnt = 0
for i in range(n, m + 1):
    if not dist[i]:
        bfs(i)
        cnt += 1
print(cnt)
print(dist)
```
