큐는 선입선출 자료구조로 제일 먼저 들어온 데이터가 가장 먼저 나가고 가장 늦게 들어온 데이터가 가장 늦게 나가는 구조   
인큐(enque) :데이터를 삽입    
디큐(deque) :데이터를 꺼내는것    
인큐에 append(), 디큐에 pop() 사용

```python
class Queue:
    def __init__(self):
        self.lists = []
        
    def enqueue(self,data):
        self.lists.append(data)
        print('push : ',data,end=' ')
        
        
    def dequeue(self):
        return self.lists.pop(0)
        
    def empty(self):
        if not self.lists:
            return True
        else:
            return False
    
    def peek(self):
        return self.lists[0]
    
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)
print('\n')

while not q.empty():
    print('pop :  ',q.dequeue(),end=' ')
```
