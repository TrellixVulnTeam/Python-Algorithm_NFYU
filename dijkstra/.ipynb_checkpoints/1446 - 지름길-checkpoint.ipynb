```python
import heapq

INF = int(1e9)
n, d = map(int, input().split())
graph = [[] for _ in range(d)]
distance = [INF] * (d)

for _ in range(n):
    a, b, c = map(int, input().split())
    graph[a].append((b, c))

def dijkstra(start):
    q = []
    heapq.heappush(q, (0, start))
    distance[start] = 0
    print(distance)
    dist, node = heapq.heappush(q)
    
    while q:
        if dist > distance[node]:
            continue

        for i in graph[node]:
            cost = dist + i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q)
    return distance
start = 0
ans = dijkstra(start)
print(ans)
```
